```python
#merging data
#two diff sources of data and you want to merge them on a particular key
person = pd.read_csv('survey_person.csv')
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
survey
```

```python
oto = person.merge(survey,left_on='ident',right_on='person')
oto
```

```python
#one to one merge
o2o = site.merge(visited,left_on='name',right_on='site')
o2o
```

```python
person.merge(survey, left_on='ident', right_on='person')
```
